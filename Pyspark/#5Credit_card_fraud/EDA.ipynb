```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.feature import VectorAssembler
```

```python
number = ["V"+ str(x) for x in range(1,29)]
```

```python
assembler = VectorAssembler()\
            .setInputCols(["Time", 'V1', 'V2',
 'V3',
 'V4',
 'V5',
 'V6',
 'V7',
 'V8',
 'V9',
 'V10',
 'V11',
 'V12',
 'V13',
 'V14',
 'V15',
 'V16',
 'V17',
 'V18',
 'V19',
 'V20',
 'V21',
 'V22',
 'V23',
 'V24',
 'V25',
 'V26',
 'V27',
 'V28',
 "Amount"])\
            .setOutputCol("features")
assembler_df= assembler.transform(df)
```

```python
assembler_df.toPandas().head()
```

```python
train,test = assembler_df.randomSplit([0.8,0.2],seed=123)
```

```python
#### LOGISTIC REGRESSION
```

```python
train_1 = train.select("features","Class")
test_1 = test.select("features","Class")
```

```python
from pyspark.ml.classification import RandomForestClassifier
```

```python
est = RandomForestClassifier()
est.setMaxDepth(5)
est.setLabelCol("Class")
```

```python
model = est.fit(train_1)
```

```python
predictions = model.transform(test)
```

```python
clf = LogisticRegression(maxIter=5,featuresCol ='features', labelCol='Class', weightCol ="Class")
```

```python
clf_model = clf.fit(train_1)
```

```python
predictions = clf_model.transform(test_1)
```

```python
accuracy = predictions.filter(predictions.Class == predictions.prediction).count() / float(predictions.count())
print("Accuracy : ",accuracy)
Accuracy :  0.7446808510638298
```

```python
from pyspark.mllib.evaluation import MulticlassMetrics
import pyspark.sql.functions as F
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_true = predictions.select(['Class']).collect()
y_pred = predictions.select(['prediction']).collect()
```

```python
print(classification_report(y_true, y_pred))
```

```python
print(confusion_matrix(y_true, y_pred))
```
